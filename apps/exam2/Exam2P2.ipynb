```python
#In wildlife biology, we need to establish the range in which a migratory predator hunts.   That is done by using a radio collar 
# that reports the location of the animal within an area.    Assume that the animal's range is contained with a square area where 
# the lower left corner is set to (0,0) and the upper right corner is set to (1000,1000).    The radio collar transmit data to a server 
# that dumps the coordinates into a file that contains one data point per line, with x and y coordinates separated by a comma.     
# Write a function in Python that is passed a filename in a string and reads this data into a list of tuples.   Return this list of tuples back to the caller.

#Hints and notes:

#A sample data file can be found in the Exam 2 folder on the course page in Blackboard.
#Look real close at the format of the file.   It's a text file in CSV format.  Remember what I said about always checking to
#  see if someone has already done the work for you.  I bet you can find a package that has functions that will help you read a 
# CSV file if you spend a few moments checking the web before answering this question.
#There is a trick to appending a tuple to a list of tuples:
#Just trying something like this will result in an error:  myList = myList + (a,b)
#You have to wrap the tuple into  a list:  myList = myList + [(a,b)]

#!! used code from: https://thispointer.com/python-read-csv-into-a-list-of-lists-or-tuples-or-dictionaries-import-csv-to-list/
from csv import reader

def readData(fileName):
    # read csv file as a list of lists
    with open(fileName, 'r') as read_obj:
        # pass the file object to reader() to get the reader object
        csv_reader = reader(read_obj)
        # Pass reader object to list() to get a list of lists
        data = list(map(tuple, csv_reader))
        print(data)

    return data

data = readData("e2p2data.csv")
```
